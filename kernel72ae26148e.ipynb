```python
df=pd.read_csv('../input/mydf.csv')
```

```python
del(df['Unnamed: 0'])
```

```python
Y_train=df['damage_grade'] 
del(df['damage_grade'])
X_train=df
```

```python
all_data=X_train
all_data_t=X_train.T
covar=np.cov(all_data_t)
eigen_val, eigen_vectors=np.linalg.eig(covar)
eigen_val_vector_pair=[]

for i in range(len(eigen_val)):
    eig_vec=eigen_vectors[:, i]
    eigen_val_vector_pair.append((eigen_val[i], eig_vec))
eigen_val_vector_pair.sort(reverse=True)
eigen_val_vector_pair


counter_eig=0
for i in (eigen_val):
    if(i>0):
        counter_eig+=1
counter_eig
```

```python
pca=decomposition.PCA(n_components=54)
x_train_pca=pca.fit_transform(X_train)
x_train, x_test, y_train, y_test=train_test_split(x_train_pca, Y_train, test_size=0.4, random_state=1)
```

```python
clf1=RandomForestClassifier(max_depth=3, min_samples_split=4, bootstrap=True, n_estimators=70, verbose=2)
```

```python
clf2 = AdaBoostClassifier(ExtraTreesClassifier(n_estimators=10, bootstrap = True, min_samples_split=2, max_depth= None, verbose=2), n_estimators=10, learning_rate=0.3, algorithm='SAMME')
```

```python
clf3= MLPClassifier(activation='tanh', alpha=3, solver='lbfgs', hidden_layer_sizes=(30,25,15), verbose=True, learning_rate='adaptive', random_state=1, tol=0.00001)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
meta=GradientBoostingClassifier(learning_rate = 0.3, n_estimators=10)
sclf=StackingClassifier(classifiers=[clf1, clf2, clf3], meta_classifier=meta, use_probas=True, average_probas=False, verbose=2)
sclf.fit(x_train, y_train)
```

```python
y_pred=sclf.predict(x_test)
accuracy_score(y_pred, y_test)
```
